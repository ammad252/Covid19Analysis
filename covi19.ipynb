***Compare Recovered Rate Between Pak vs India***

```python
pak_recovery_rate = (pakistan_data["Recovered"].max() / pakistan_data["Confirmed"].max()) * 100
india_recovery_rate = (india_data["Recovered"].values[0] / india_data["Confirmed"].values[0]) * 100
print("Pakistan Recovery Rate:", round(pak_recovery_rate, 2), "%")
print("India Recovery Rate:", round(india_recovery_rate, 2), "%")


if pak_recovery_rate > india_recovery_rate:
    print("Pakistan ka recovery rate zyada hai ✅")
elif india_recovery_rate > pak_recovery_rate:
    print("India ka recovery rate zyada hai ✅")
else:
    print("Both  are Good ✅")


```

```python
region_summary = covid_datasets.groupby("WHO Region")[["Confirmed last week", "1 week change", "1 week % increase"]].sum()

print(region_summary)
```

```python
top10_new_deaths = covid_datasets.nlargest(10, "New deaths")[["Country/Region", "New deaths"]]
print(top10_new_deaths)
```

```python
plt.figure(figsize=(12,6))
plt.bar(top10_new_deaths["Country/Region"], top10_new_deaths["New deaths"], color="red")

plt.xlabel("Country/Region")
plt.ylabel("New Deaths")
plt.title("Top 10 Countries by New Deaths")
plt.show()
```

```python
top10_new_recovered = covid_datasets.nlargest(10, "New recovered")[["Country/Region", "New recovered"]]
print(top10_new_recovered)
```

```python
plt.figure(figsize=(12,6))
plt.bar(top10_new_recovered["Country/Region"], top10_new_recovered["New recovered"], color="red")

plt.xlabel("Country/Region")
plt.ylabel("New recovered")
plt.title("Top 10 Countries by New Recovered")
plt.show()
```
